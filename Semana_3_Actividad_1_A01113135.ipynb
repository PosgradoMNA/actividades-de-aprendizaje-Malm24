# **Fundamentos de bases de datos y para ciencia de datos.**

 La información puede ser recopilada en distintas formas y estructura:

Datos no estructurados: Documentos de texto, imagenes y audios.
Datos estructurada: datos que se encuentran organizados en formato de tabla.

Actualmente SQL domina la industria de los datos estructurados y es el sistema de bases de datos que normalmente se encuentra en las empresas.

Una base de datos es una tabla con indentificador de filas y encabezados para sus columnas. Estas tablas almacenan distintos tipo de datos y en diferente niveles de detalle.

A las filas de una base de datos regularmente se les llama record. Mientras que a las columnas se les puede llamar atributos o campos.

Primary Key es el nombre asignado a un valor unico que permite indentificar una fila.

Foreign key es cuando utilizamos una primary key  como referencia en una nueva tabla.

Normalización de base de datos: es el concepto utilizado para evitar el registro de datos redundantes y así reducir la cantidad de datos almacenados.

# **Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.**

Un Data Warehouse puede ser descrito como un base de datos gigante que integra información de distintas fuentes.  

El mayor reto de un data warehouse es tomar la información de distintas fuentes y habilitar que las personas que consumen dicha información puedan consultar diversas fuentes de manera simultánea. Para lograr esto es necesario lidiar con información faltante, mantener las escalas, terminología y otros temas importantes.

ETL consiste en transformar un grupo de data de manera offline para posteriormente importarla y cargarla en un data warehouse. Por su parte ELT consiste en extraer información cruda, cargarla en un sistema data warehouse y utilizar el poder del mismo para transformar la información.
