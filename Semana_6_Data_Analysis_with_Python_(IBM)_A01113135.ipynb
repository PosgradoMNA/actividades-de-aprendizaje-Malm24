##Análisis Exploratorio de Datos

Es un paso preeliminar en el análisis de datos y permite resumir las caracteristicas de los datos, entender el data set, identificar la relación entre variables y extraer variables importantes

df.describe() nos permite conocer una análisis general de nuestros datos (total de datos, promedio, desviación estandar, min, max y cuartiles)

values_counts() nos permite sumar los datos categoricos.

Un boxplot nos permite observar la distribución de los datos utilizando min,max, cuartiles, mediana y datos atipicos. Este gráfico nos permite comparar la distribución de distintos datos.

Scatter Plot nos muestra la relación entre dos variables

eje x - variable independiente
eje y - variable dependiente

para crear este gráfico debemos definir los ejes ("X" y "Y")
plt.scatter(x,y)

para dar un título al gráfico y su ejes utilizamos el siguiente código:

plt.title > definir título de gráfico 
plt.xlabel > definir título de eje x
plt.ylabel > definir título de eje y

groupby() - se aplica para variables categorica y permite clasificar los datos por grupos.

pivot.table() - nos permite crear una tabla con una variable en las filas y otra en columnas y analizar el promedio de estos grupos.

Heatmap nos permite observar el comportamiento por medio de colores de la variable de interes comparado utilizando una variable independiente en el eje "X" y otra en el eje "Y).


##ANOVA - Análisis de varianz  a - stats.f_oneway()

Permite encontrar la correlación entre diferentes grupo de variables categoricas

Este análisis nos entrega un f-test scores y un p-value.

F-test score: variación entre la media de los grupos dividido entre la variación entre los grupos.

p-value: grado de confianza


## Correlación - sns.regplot(x="ex1, y"ex2", data=df)


Indentificar que variables son interdependientes.

 La relaciones pueden ser positiva o neogativas

Correlación no implica siempre una causalidad. 

## Metodo de correlación de Pearson 

pearso_ coef, p_value = stats.pearsonr(df[],df[])

Este análisis nos entrega una coeficiente de relación y un p-value

Interpetación coeficiente de relación: 
Cercano a +1 -> Relación positiva fuerte
Cercano a 0  -> no relación
Cercano a -1 -> Relación negativa fuerte

interpretación p-value:

0.001 - certeza fuerte
0.05 -  certeza moderada
p<0.1 - certeza debil 
p>0.1 - no hay certeza

```python
df_gptest = df[['drive-wheels','body-style','price']] #crean un nuevo df
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean() #calculando el promedio de los grupos con variables multiples
grouped_test1
```

```python
grouped_pivot = grouped_pivot.fillna(0) #llenando los valores vacios con 0
grouped_pivot
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
plt.pcolor(grouped_pivot, cmap='RdBu') #creando gráfico ip
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#nombre de etiquetas
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#moviendo etiquetas al centro
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insertando etiquetas
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotando etiquetas
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

```python
grouped_test2=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels']) #creando un grupos de data para conocer precio promedio
grouped_test2.head(2)
```

```python
grouped_test2.get_group('4wd')['price']
```

```python
# realizando  ANOVA entre tres grupos
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

El valor nos indica que la diferencia entre estos grupo es sigifnicativa y el p-value tan pequeño nos indica un bajo grado de confianza

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'])  ## realizando  ANOVA entre dos grupos
 
print( "ANOVA results: F=", f_val, ", P =", p_val )
```

El valor nos indica que la diferencia entre estos grupo es sigifnicativa y el p-value tan pequeño nos indica un alto grado de confianza
